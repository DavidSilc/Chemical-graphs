```python
from sage.all import *

def degree_configuration(graph):
    """
    Funkcija vrne konfiguracijo grafa.

    Arguments:
        graph: Sagemath graf objekt

    Returns:
        tuple: seznam oblike (a1, a2, a3, a4) , kjer:
            a1 = število vozlišč s stopnjo 1
            a2 = število vozlišč s stopnjo 2
            a3 = število vozlišč s stopnjo 3
            a4 = število vozlišč s stopnjo 4
    """
    degree_counts = [0, 0, 0, 0]
    for degree in graph.degree():
        degree_counts[degree - 1] += 1
    return tuple(degree_counts)

def generate_unique_chemical_graphs_configs(n):
    """
    Funkcija generira kemijske grafe z unikatnimi konfiguracijami.

    Args:
        n (int): Število vozlišč grafa

    Returns:
        list: Seznam vseh unikatnih konfiguracij
    """
    unique_graphs = []
    for g in graphs.nauty_geng(f"{n} -c -D4"):
        config = degree_configuration(g)
        if config not in unique_graphs:
            unique_graphs.append(config)
    return unique_graphs


    
def sigma_total_irregularity_from_config(config, f):
    """
    Funkcija izračuna sigma totalni indeks iz konfiguracije.

    Args:
        tuple: Seznam (a1, a2, a3, a4)

    Returns:
        int: Sigma totalni indeks
    """
    sigma_t = 0
    f_n = f(n)
    a1, a2, a3, a4 = config
    degree_list = [1] * a1 + [2] * a2 + [3] * a3 + [4] * a4
    
    sigma_t = 0
    for i in range(len(degree_list)):
        for j in range(i + 1, len(degree_list)):
            sigma_t += abs(degree_list[i] - degree_list[j]) ** f_n

    return round(sigma_t,3)
```

```python
def ena_z_n(n):
    f = lambda n: 1/n
    config_with_sigma = [(config, sigma_total_irregularity_from_config(config, f)) for config in chemical_graphs]
    config_with_sigma.sort(key=lambda x: x[1], reverse=True)
    print(f"Vse unikatne konfiguracije kemičnih grafov na {n} vozliščih, urejenih padajoče po sigma-totalnem indeksu za funkcije f(n) = 1/n = {round(1/n,2)} :")
    for config, sigma in enumerate(config_with_sigma):
        print(config, sigma)
```
