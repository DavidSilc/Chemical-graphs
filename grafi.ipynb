```python
# (a) Generacija kemijskih grafov

def generate_chemical_graphs(n, a1, a2, a3, a4):
    """
    Ustvari vse kemijske grafe za dano n in podano stopenjsko zaporedje (a1, a2, a3, a4).
    """
    from sage.graphs.graph_generators import graphs

    # Preverimo, ali je konfiguracija smiselna
    if a1 + a2 + a3 + a4 != n:
        raise ValueError("Vsota a1, a2, a3, a4 mora biti enaka n.")

    chemical_graphs = []

    # Generiramo vse grafe s natanko n vozlišči
    for G in graphs(n):
        # Preverimo, ali graf izpolnjuje pogoje kemijskega grafa
        degrees = G.degree()
        if all(d <= 4 for d in degrees):  # Maksimalna stopnja 4
            # Preštejmo število vozlišč z določeno stopnjo
            counts = [degrees.count(i) for i in range(1, 5)]
            if counts == [a1, a2, a3, a4]:
                chemical_graphs.append(G)

    return chemical_graphs

# (b) Izračun sigma vrednosti

def sigma_f_t(G, f):
    """
    Izračuna sigma_f_t(G) za graf G in funkcijo f.
    """
    from itertools import combinations

    sigma_value = 0

    # Iteriramo čez vse pare vozlišč
    for u, v in combinations(G.vertices(), 2):
        degree_difference = abs(G.degree(u) - G.degree(v))
        sigma_value += degree_difference ** f(len(G.vertices()))

    return sigma_value

# Primeri funkcij f(n)
def f_n_inverse(n):
    return 1 / n

def f_n_constant(n, c=0.5):
    return c

# Uporaba funkcij
if __name__ == "__main__":
    # Parameterji za generacijo grafov
    n = 8  # Število vozlišč
    a1, a2, a3, a4 = 2, 2, 2, 2  # Stopenjsko zaporedje (a1, a2, a3, a4)

    # Generacija kemijskih grafov
    print("Generacija kemijskih grafov...")
    graphs = generate_chemical_graphs(n, a1, a2, a3, a4)
    print(f"Najdenih {len(graphs)} grafov.")

    # Izračun sigma vrednosti za vsakega
    print("Izračun sigma vrednosti za f(n) = 1/n...")
    for i, G in enumerate(graphs):
        sigma_value = sigma_f_t(G, f_n_inverse)
        print(f"Graf {i + 1}: Sigma = {sigma_value}")

    # Izračun sigma vrednosti za f(n) = c (npr. c = 0.5)
    print("Izračun sigma vrednosti za f(n) = 0.5...")
    for i, G in enumerate(graphs):
        sigma_value = sigma_f_t(G, lambda n: f_n_constant(n, c=0.5))
        print(f"Graf {i + 1}: Sigma = {sigma_value}")

    # Risanje grafov
    print("\nRisanje grafov...")
    for i, G in enumerate(graphs):
        print(f"Risanje grafa {i + 1}...")
        G.show(title=f"Graf {i + 1}")

    # Uporaba za n = 7
    print("\nUporaba za n = 7...")
    n = 7
    a1, a2, a3, a4 = 2, 1, 2, 2  # Spremenite konfiguracijo po potrebi

    graphs = generate_chemical_graphs(n, a1, a2, a3, a4)
    print(f"Najdenih {len(graphs)} grafov za n = {n} in konfiguracijo (a1, a2, a3, a4) = ({a1}, {a2}, {a3}, {a4}).")

    for i, G in enumerate(graphs):
        sigma_value = sigma_f_t(G, f_n_inverse)
        print(f"Graf {i + 1}: Sigma = {sigma_value}")

    # Risanje grafov
    print("\nRisanje grafov...")
    for i, G in enumerate(graphs):
        print(f"Risanje grafa {i + 1}...")
        G.show(title=f"Graf {i + 1}")
```
